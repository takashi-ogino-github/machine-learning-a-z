# Data Preprocessing Tools

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```


## Importing the dataset

- colab上で毎回driveをマウントしてこのファイルが存在する場所に行く必要があるっぽい

```python
dataset = pd.read_csv('Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values 
```

```python
print(X)
```

```python
print(y)
```

## Taking care of missing data

```python
from sklearn.impute import SimpleImputer # sklearnのimputeモジュールからSimpleImputerクラスをimport
imputer = SimpleImputer(missing_values=np.nan, strategy='mean') # SimpleImputerクラスのオブジェクトを作成
imputer.fit(X[:, 1:3]) # imputeオブジェクトをfit methodによってデータセットに適用, データセットの数値カラムのすべての行を指定
X[:, 1:3] = imputer.transform(X[:, 1:3])
```

```python
print(X)
```

## Encoding categorical data

### Encoding the Independent Variable

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough') # transform: どのような変換を、どのようなクラスで、どの列のindexに施すか、remainder: 変換を施さない列をどうするかを指定
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

### Encoding the Dependent Variable

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(y)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:, 3:] = sc.fit_transform(X_train[:, 3:]) # train setを用いて標準化関数を計算 & 適用
X_test[:, 3:] = sc.transform(X_test[:, 3:]) # train setと同じ標準化関数で標準化を適用
```
