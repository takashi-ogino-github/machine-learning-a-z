# Simple Linear Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression # 単回帰用のクラスをimport
regressor = LinearRegression() # 単回帰モデルのオブジェクトを生成
regressor.fit(X_train, y_train) # fitメソッドで学習, trainデータセットを用いる
```

## Predicting the Test set results

```python
y_pred =  regressor.predict(X_test)
```

## Visualising the Training set results

```python
plt.scatter(X_train, y_train, color = 'red') # train dataset（実測値）を散布図として描画
plt.plot(X_train, regressor.predict(X_train), color = 'blue') # 訓練したモデルによる回帰直線を描画
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Test set results

```python
plt.scatter(X_test, y_test, color = 'red') # test dataset（実測値）を散布図として描画
plt.plot(X_train, regressor.predict(X_train), color = 'blue') # 訓練したモデルによる回帰直線を描画
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
